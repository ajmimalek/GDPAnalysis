## DataSet Creation
The DataSet chosen is imported from this link https://www.kaggle.com/zackerym/gdp-annual-growth-for-each-country-1960-2020

```python
import pandas as pd
data=pd.read_csv("C:/Users/Malek/Desktop/Révision/ML/GDP_annual_growth.csv",header=0,sep =',')
```

```python
data.head()
```

The DataSet presents GDP for every country uin the world and the variation of the GDP through the years. <br/>
We are going to work only on Tunisia

```python
data = data.loc[data["Country Name"] == "Tunisia"]
```

```python
data.shape
```

```python
data.head()
```

```python
#We are going to delete none values
data = data.dropna(axis=1)
```

```python
#Importing the plot library
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
#Getting GDPs of different years
data_years = data.iloc[:,4:]
data_years.head()
```

```python
data.columns
```

```python
#Getting GDP Summary
print("Tunisia GDP Summary")
print()
print('Mean : ', end= '')
print(data_years.mean(axis=1))
print('Total GDP Until Now: ', end='')
print(data_years.sum(axis=1))
print("Maximum GDP: ", data_years.max(axis=1))
print("Year of the Maximum GDP: ", data_years.idxmax(axis=1))
print("Minimum GDP: ", data_years.min(axis=1))
print("Year of the Minimum GDP: ", data_years.idxmin(axis=1))
```

```python
#A graph showing the variation of GDPs through the years
plt.figure(figsize=(30,22))
plt.title(label= 'Tunisia GDP Analysis', fontsize=30, fontweight=20)
plt.ylabel('GDP in terms of $')
plt.xlabel('Years',fontsize=25)
plt.plot(np.arange(1965,2020),data_years.to_numpy()[0])
plt.xticks(np.arange(1965,2020),np.arange(1965,2020), rotation=45)
plt.grid()
```
